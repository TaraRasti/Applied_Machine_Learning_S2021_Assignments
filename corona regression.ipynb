```python
import numpy as np
import pandas as pd
from glob import glob
filenames=glob(r'C:\Users\Dell\Desktop\COVID-19-master\csse_covid_19_data\csse_covid_19_daily_reports\*.csv')
dataList=[pd.read_csv(f) for f in filenames]
```

```python
selected_featurs_list=[]
for i,df in enumerate(dataList):
  dayDf=pd.DataFrame(pd.Series([i]).repeat(df.shape[0]),columns=['day'])
  dayDf.index=range(df.shape[0])
  selDf=df[['Country/Region','Province/State','Confirmed']]
  selected_featurs_list.append(dayDf.join(selDf))
```
